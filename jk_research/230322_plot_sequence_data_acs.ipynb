```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
sns.set_style("dark")
# also increase the font size dramatically, and make the plot slightly bigger
sns.set(font_scale=2.4)
plt.rcParams["figure.figsize"] = (8, 8)
```

```python
num_uniqe = 164778
num_rcsb = 202764
num_uniprot = 245871724
```

```python
# Now make the same chart but with a log scale
# First, make a dataframe with the data
data = pd.DataFrame(
    {
        
        "Number of sequences": [num_rcsb, num_uniqe, num_uniprot],
        "Source": ["RCSB Structures", "RCSB Structures\nunique", "UniProt\nSequences"],
    }
)
# Then, make the plot, color the bars tab:blue and tab:orange
sns.barplot(
    x="Number of sequences",
    y="Source",
    data=data,
    palette=["tab:blue", "tab:orange", "tab:green"],
    linewidth=2,
)
plt.xlabel("Dataset Size")
plt.ylabel("")
plt.xscale("log")
plt.tight_layout()
# Only show the last tick on the x axis, and use latex to make it look nice
plt.xticks([], [])

# Now add labels showing the number of sequences
plt.text(
    num_rcsb + 1e6,
    0,
    f"{num_rcsb:,}",
    ha="left",
    va="center",
    color="black",
    fontsize=22,
)
plt.text(
    num_uniqe + 1e6,
    1,
    f"{num_uniqe:,}",
    ha="left",
    va="center",
    color="black",
    fontsize=22,
)
# Plot the text for the last bar on top of the bar itself
plt.text(
    num_uniprot + 1e6,
    2,
    f"{num_uniprot:,}",
    ha="left",
    va="center",
    color="black",
    fontsize=22,
)



plt.tight_layout()
plt.savefig("sequence_data_summary_log.pdf")
plt.show()
```
