When I created the validation set at first, the script mistakenly did not exclude proteins
with sequences longer than 700. This modifies the validation data directory to exclude
these proteins.

```python
long_proteins = []
with open("../data/validation/cameo/20220116/input.fasta", mode='r') as f, open("long_proteins.txt", mode='w') as g:
    title = ""
    seq = ""
    for line in f:
        if line.startswith(">"):
            title = line[1:].strip()
        else:
            seq = line.strip()
            if len(seq) > 700:
                print(title)
                print(seq)
                long_proteins.append(title)
                g.write(title + "\n")
           
```

```python
# Use bash and grep with iPython magic to remove the long proteins from the fasta file
!grep -v -f long_proteins.txt ../data/validation/cameo/20220116/input.fasta > ../data/validation/cameo/20220116/input_no_long_proteins.fasta
!grep -v -f long_proteins.txt ../data/validation/cameo/20220116/fastalist.txt > ../data/validation/cameo/20220116/fastalist2.txt

# Do the same to the datalist.txt file, but only look at the first 4 characters of the long_proteins list
!grep -v -f <(cut -c 1-4 long_proteins.txt) ../data/validation/cameo/20220116/datalist.txt > ../data/validation/cameo/20220116/datalist2.txt
```

```python
# Long protein should only be in original files
!grep 7b9p ../data/validation/cameo/20220116/*txt
!grep 7b9p ../data/validation/cameo/20220116/*fasta
```

```python
# Iterate through the long proteins and move the matching directories under ../data/validation/cameo/20220116/alignments to ../data/validation/cameo/20220116/alignments_too_long using iPython magic
for protein in long_proteins:
    print(protein)
    !mv ../data/validation/cameo/20220116/alignments/$protein ../data/validation/cameo/20220116/alignments_too_long

# Now do the same for the data files in data_dir
data_dir = "../data/validation/cameo/20220116/data_dir/"
data_dir_too_long = "../data/validation/cameo/20220116/data_dir_too_long/"
for protein in long_proteins:
    pshort = protein[:4]
    # print(protein[:4])
    !echo "$pshort".cif
    !mv "$data_dir$pshort".cif "$data_dir_too_long$pshort".cif

# Now do the same for the fasta files in fasta_dir
fasta_dir = "../data/validation/cameo/20220116/fasta_dir/"
fasta_dir_too_long = "../data/validation/cameo/20220116/fasta_dir_too_long/"
!mkdir "$fasta_dir_too_long"
for protein in long_proteins:
    pshort = protein
    # print(protein[:4])
    !echo "$pshort".fasta
    !mv "$fasta_dir$pshort".fasta "$fasta_dir_too_long$pshort".fasta
```
